```python
import pandas as pd
import numpy as np
```

```python
datos = pd.read_excel('Temperatura Maxima y Minima Promedio Mensual de Coquimatlán.xlsx')
datos
```

```python
#Promedio de las temperaturas maximas
datos['MEDIA'].mean()
```

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
estaciones = ['1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']
temperaturas_medias_Max_Anual = [39.132, 39.579, 34.944, 33.753, 32.449, 31.357, 33.822, 33.612, 33.654, 33.678, 33.551, 34.013, 34.267, 34.230, 33.849, 34.027, 34.508, 34.496, 33.847, 34.474, 33.316, 34.114, 32.488, 32.410, 33.231, 33.385, 33.459, 33.050]

# Crear gráfico de barras para las temperaturas medias
plt.bar(estaciones, temperaturas_medias_Max_Anual)

# Rotar las etiquetas del eje x
plt.xticks(rotation=45, ha='right')

plt.xlabel('Estaciones')
plt.ylabel('Temperatura Media (°C)')
plt.title('Temperaturas Medias por Estación')

# Mostrar el gráfico
plt.show()
```

```python
#Sacar la moda
moda = datos['MEDIA'].mode
print("la moda es : ",moda)
```

```python
#Sacar la desviacion estandar
desviacion_estandar = datos['MEDIA'].std()
print("la desviacion estandar es de: ",desviacion_estandar)
```
