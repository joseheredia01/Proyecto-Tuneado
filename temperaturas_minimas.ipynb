```python
import matplotlib.pyplot as plt

# Datos de ejemplo
estaciones = ['1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']
temperaturas_medias_Min_Anual = [18.042, 17.475, 17.837, 16.658, 17.534, 17.787, 17.440, 19.372, 19.257, 18.893, 19.363, 19.307, 19.726, 19.842, 19.661, 19.831, 20.169, 20.013, 19.119, 20.089, 19.569, 20.182, 20.058, 20.240, 20.482, 20.796, 20.324, 20.254]

# Crear gráfico de barras para las temperaturas medias
plt.bar(estaciones, temperaturas_medias_Min_Anual)

# Rotar las etiquetas del eje x
plt.xticks(rotation=45, ha='right')

plt.xlabel('Estaciones')
plt.ylabel('Temperatura Media (°C)')
plt.title('Temperaturas Medias por Estación')

# Mostrar el gráfico
plt.show()
```

```python
#Sacar la moda
moda = datos['MEDIA'].mode
print("la moda es : ",moda)
```
